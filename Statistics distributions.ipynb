Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

Answer: 

```python
from scipy.stats import norm 
import numpy as np 
  
data_start = -5
data_end = 5
data_points = 11
data = np.linspace(data_start, data_end, data_points) 
  
mean = np.mean(data) 
std = np.std(data) 
  
probability_pdf = norm.pdf(3, loc=mean, scale=std) 
print(probability_pdf) 
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

Answer: 

A binomial experiment is one that has the following properties: 

>(1) The experiment consists of n identical trials. 

>(2) Each trial results in one of the two outcomes, called a success S and failure F. 

>(3) The probability of success on a single trial is equal to p and remains the same from trial to trial.

A few circumstances where we have binomial experiments are,

*tossing a coin: head or tail, 

*the result of a test: pass or fail, 

*elected in an interview: yes/ no, or 

*nature of the product: defective/non-defective. 

such a distribution of a binomial random variable is called a binomial probability distribution.

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

Answer:

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

# Parameters
n = 1000  # Number of trials
p = 0.4  # Probability of success
sample_size = 1000  # Size of the random sample

# Generate a random sample from a binomial distribution
random_sample = np.random.binomial(n, p, sample_size)

# Plot a histogram of the results
plt.hist(random_sample, bins=np.arange(-0.5, n + 1.5, 1), density=True, alpha=0.75, color='b', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p}) - Sample Size: {sample_size}')
plt.xticks(range(n + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the histogram
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset('taxis')

# Extract the 'hour' and 'fare' columns
data['hour'] = data['pickup'].dt.hour
df = data[['hour', 'fare']]

# Group data by 'hour' and calculate the mean 'fare' for each hour
hourly_means = df.groupby('hour')['fare'].mean()

# Calculate the Poisson CDF
from scipy.stats import poisson

# Set the parameters for the Poisson distribution
poisson_lambda = hourly_means  # Using hourly mean fares as the Poisson parameter

# Calculate the Poisson CDF for a specific point (e.g., hour 12)
specific_hour = 12
cdf_at_specific_hour = poisson.cdf(specific_hour, poisson_lambda)

# Plot the Poisson CDF for hour 12
x = np.arange(0, 30)  # Set the x-values
cdf_values = poisson.cdf(x, poisson_lambda[specific_hour])
plt.plot(x, cdf_values, marker='o')
plt.title(f'Poisson CDF for Hour {specific_hour}')
plt.xlabel('Number of Events')
plt.ylabel('CDF')
plt.grid(True)
plt.show()

print(f'Poisson CDF at hour {specific_hour}: {cdf_at_specific_hour}')
```

Q7. How Binomial distribution different from Poisson distribution?

Answer:
    
    
The Binomial and Poisson distributions deal with random variables and probabilities, they differ in the nature of the experiment, number of trials, probability of success, and probability distribution.

>Nature of the experiment: The Binomial distribution applies to experiments with a fixed number of trials, while the Poisson distribution applies to events occurring within a fixed interval.


>Number of trials: The Binomial distribution requires a fixed number of trials, whereas the Poisson distribution does not have a predetermined number of events.


>Probability of success: In the Binomial distribution, the probability of success remains constant for each trial, whereas the Poisson distribution does not explicitly consider the probability of success.


>Probability distribution: The Binomial distribution represents the probability of achieving a specific number of successes, while the Poisson distribution models the probability of observing a certain number of events.
