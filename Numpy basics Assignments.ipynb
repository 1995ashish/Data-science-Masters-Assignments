```python
import numpy as np
list_ = ['1','2','3','4','5']
array_list = np.array(object = list_)
```

```python
type(list_)
```

```python
type(array_list)
```

Answer: The data types are different as the list_ is the List data type and the array_list data type is array.

```python
for i in range(0,5):
    a=list_[i]
    print(type(a))
```

```python
for i in range(0,5):
    a=array_list[i]
    print(type(a))
```

```python
array_list = np.array(object = list_, dtype = int)
```

```python
for i in range(0,5):
    a=array_list[i]
    print(type(a))
```

```python
for i in range(0,5):
    a=list_[i]
    print(type(a))
```
